```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
import matplotlib.pyplot as plt
import matplotlib_inline
matplotlib_inline.backend_inline.set_matplotlib_formats('svg')
```

```python
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
import seaborn as sns
iris = sns.load_dataset('iris')

iris.head()
```

```python
# organise the data

# convert from pandas df to tensor
data = torch.tensor(iris[iris.columns[0:4]].values).float()

# transform species to numbers
labels = torch.zeros(len(data), dtype=torch.long)
# labels[iris.species == 'setosa'] = 0 not required as already zeroes are there
labels[iris.species == 'versicolor'] = 1
labels[iris.species == 'virginica'] = 2
```

```python
class createModel(nn.Module):
    def __init__ (self, nUnits, nLayers):
        super().__init__()

        self.nUnits = nUnits
        self.nLayers = nLayers

        self.layers = nn.ModuleDict()

        self.layers['input'] = nn.Linear(4, nUnits)

        for n in range(self.nLayers):
            self.layers[f'hidden-{n}'] = nn.Linear(self.nUnits, self.nUnits)

        self.layers['output'] = nn.Linear(nUnits, 3)
        
    def forward(self, data):
        x = F.relu(self.layers['input'](data))

        for n in range(self.nLayers):
            x = F.relu(self.layers[f'hidden-{n}'](x))
        
        x = self.layers['output'](x)

        return x

        
```

```python
# just testing the class
testModel = createModel(1, 1)
print(testModel)
testModel.forward(torch.randn(10,4)).shape
```

```python
epochs = 500
def trainModel(model):
    lossFunction = nn.CrossEntropyLoss()
    optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
    for epoch in range(epochs):
        yHat = model(data)

        loss = lossFunction(yHat, labels)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    
    predictions = model(data)
    accuracy = torch.mean((torch.argmax(predictions, axis=1) == labels).float())*100
    nParams = sum([ p.numel() for p in model.parameters() if p.requires_grad ])
    return accuracy, nParams
```

```python
# just testing the function
testFun = trainModel(testModel)
testFun
```

```python
# actual test

numLayers = range(1, 6)
numUnits = np.arange(4, 101, 3)

accuracies = np.zeros((len(numUnits), len(numLayers)))
params = np.zeros((len(numUnits), len(numLayers)))


for unit in range(len(numUnits)):
    for layer in range(len(numLayers)):
        print(f'units -> {numUnits[unit]}')
        print(' ')
        print(f'layers -> {numLayers[layer]}')
        model = createModel(numUnits[unit], numLayers[layer])
        acc, nparams = trainModel(model)

        accuracies[unit, layer] = acc
        params[unit, layer] = nparams
```

```python
accuracies
```

```python
fig,ax = plt.subplots(1,figsize=(12,6))

ax.plot(numUnits,accuracies,'o-',markerfacecolor='w',markersize=9)
ax.plot(numUnits[[0,-1]],[33,33],'--',color=[.8,.8,.8])
ax.plot(numUnits[[0,-1]],[67,67],'--',color=[.8,.8,.8])
ax.legend(numLayers)
ax.set_ylabel('accuracy')
ax.set_xlabel('Number of hidden units')
ax.set_title('Accuracy')
plt.show()
```

```python
# vectorize for convenience
x = params.flatten()
y = accuracies.flatten()

# correlation between them
r = np.corrcoef(x,y)[0,1]

# scatter plot
plt.plot(x,y,'o')
plt.xlabel('Number of trainable parameters')
plt.ylabel('Accuracy')
plt.title('Correlation: r=' + str(np.round(r,3)))
plt.show()
```
