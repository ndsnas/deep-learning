```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, TensorDataset

import numpy as np
import matplotlib.pyplot as plt
import matplotlib_inline
matplotlib_inline.backend_inline.set_matplotlib_formats('svg')

from sklearn.model_selection import train_test_split 
```

```python
# import and process data
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
import seaborn as sns
iris = sns.load_dataset('iris')

iris.head()
```

```python
# organise the data

# convert from pandas df to tensor
data = torch.tensor(iris[iris.columns[0:4]].values).float()

# transform species to numbers
labels = torch.zeros(len(data), dtype=torch.long)
# labels[iris.species == 'setosa'] = 0 not required as already zeroes are there
labels[iris.species == 'versicolor'] = 1
labels[iris.species == 'virginica'] = 2
```

```python
train_data, test_data, train_labels, test_labels = train_test_split(data, labels, train_size=0.8)

train_data = TensorDataset(train_data, train_labels)
test_data = TensorDataset(test_data, test_labels)
# test_data.tensors[0].shape[0]
batch_size = 16
train_loader = DataLoader(train_data, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_data, batch_size=test_data.tensors[0].shape[0])
```

```python
def createModel():
    return nn.Sequential(
        nn.Linear(4, 64),
        nn.ReLU(),
        nn.Linear(64, 64),
        nn.ReLU(),
        nn.Linear(64, 3)
    )
```

```python
def createANewModel():
    ANN = createModel()
    lossFunction = nn.CrossEntropyLoss()
    optimizer = torch.optim.SGD(ANN.parameters(), lr=0.005)
    return ANN, lossFunction, optimizer
```

```python
# train the model

epochs = 1000

def trainModel(ANN, lossFunction, optimizer, L1Lambda):
    trainAcc = [] # this'd contain accuracies after all batches complete, basically 1 epoch
    testAcc = []  # this'd also contain accuracies after each epoch but on test data, (the last epoch accuracy ~= final accuracy??)
    losses = []   # to store mean batch losses, similar to trainAcc

    nweights = 0  # to store total number of trainable weights in the model
    for pname, weight in ANN.named_parameters():
        if 'bias' not in pname:
            nweights += weight.numel()

    for _ in range(epochs):
        batchAcc = []
        batchLoss = []
        for x, y in train_loader:
            
            yHat = ANN(x)
            loss = lossFunction(yHat, y)

            L1_term = torch.tensor(0.0, requires_grad=True)

            for pname, weight in ANN.named_parameters():
                if 'bias' not in pname:
                    L1_term = L1_term + torch.sum(abs(weight))

            # add to loss term
            loss = loss + L1Lambda*L1_term/nweights

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            batchAcc.append((torch.mean((torch.argmax(yHat, axis=1) == y).float()) * 100).item())
            batchLoss.append(loss.item())

        trainAcc.append(np.mean(batchAcc))
        losses.append(np.mean(batchLoss))
        test_x, test_y = next(iter(test_loader))
        predLabels = ANN.forward(test_x)
        testAcc.append((torch.mean((torch.argmax(predLabels, axis=1) == test_y).float()) * 100).item())

    return trainAcc, losses, testAcc
```

```python
# create a 1D smoothing filter
def smooth(x,k=5):
  return np.convolve(x,np.ones(k)/k,mode='same')
```

### Actual experiment

```python
l1Lambdas = np.linspace(0, 0.005, 10)
allTrainAcc = np.zeros((len(l1Lambdas), epochs))
allTestAcc = np.zeros((len(l1Lambdas), epochs))
for i in range(len(l1Lambdas)):
    ANN, lossFunction, optimizer = createANewModel()
    trainAcc, losses, testAcc = trainModel(ANN, lossFunction, optimizer, l1Lambdas[i])
    allTrainAcc[i,:] = smooth(trainAcc, 10)
    allTestAcc[i,:] = smooth(testAcc, 10)

```

```python
fig, ax = plt.subplots(1,2, figsize=(12,4))

ax[0].plot(allTrainAcc.T)
ax[0].set_title('Train accuracy')
ax[1].plot(allTestAcc.T)
ax[1].set_title('Test accuracy')

# make the legend easier to read
leglabels = [np.round(i,2) for i in l1Lambdas]

for i in range(2):
    ax[i].legend(leglabels)
    ax[i].set_xlabel('Epoch')
    ax[i].set_ylabel('Accuracy (%)')
    ax[i].set_ylim([60,101])
    ax[i].grid()
plt.show()
```

```python
# show average accuracy by L1 rate

# average only some epochs
epoch_range = [10,200]


plt.plot(l1Lambdas,
         np.mean(allTrainAcc.T[epoch_range[0]:epoch_range[1],:],axis=0),
         'bo-',label='TRAIN')

plt.plot(l1Lambdas,
         np.mean(allTestAcc.T[epoch_range[0]:epoch_range[1],:],axis=0),
         'rs-',label='TEST')

plt.xlabel('L1 regularization amount')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
