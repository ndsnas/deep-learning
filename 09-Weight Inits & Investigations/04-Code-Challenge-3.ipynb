```python
import torch
import torch.nn as nn
import copy
import numpy as np
import matplotlib.pyplot as plt
import matplotlib_inline
matplotlib_inline.backend_inline.set_matplotlib_formats('svg')
```

```python
ANN = nn.Sequential(
    nn.Linear(2, 8),
    nn.Linear(8, 1),
    nn.Linear(1, 1)
)
```

```python
net_noseed = copy.deepcopy(ANN)

weights_net_noseed = np.array([])

for p in net_noseed.named_parameters():
    if 'weight' in p[0]:
        nn.init.xavier_normal_(p[1].data)
        weights_net_noseed = np.concatenate((weights_net_noseed, p[1].data.numpy().flatten()), axis=0)
        
```

```python
torch.manual_seed(1)

net_rs1a = copy.deepcopy(ANN)

weights_net_rs1a = np.array([])

for p in net_rs1a.named_parameters():
    if 'weight' in p[0]:
        nn.init.xavier_normal_(p[1].data)
        weights_net_rs1a = np.concatenate((weights_net_rs1a, p[1].data.numpy().flatten()), axis=0)
```

```python
torch.manual_seed(2)

net_rs2 = copy.deepcopy(ANN)

weights_net_rs2 = np.array([])

for p in net_rs2.named_parameters():
    if 'weight' in p[0]:
        nn.init.xavier_normal_(p[1].data)
        weights_net_rs2 = np.concatenate((weights_net_rs2, p[1].data.numpy().flatten()), axis=0)
```

```python
torch.manual_seed(1)

net_rs1b = copy.deepcopy(ANN)

weights_net_rs1b = np.array([])

for p in net_rs1b.named_parameters():
    if 'weight' in p[0]:
        nn.init.xavier_normal_(p[1].data)
        weights_net_rs1b = np.concatenate((weights_net_rs1b, p[1].data.numpy().flatten()), axis=0)
```

```python
fig = plt.figure(figsize=(15,5))
plt.plot(weights_net_noseed, 'ro', label = 'no_seed')
plt.plot(weights_net_rs1a, 'ks', label = 'rs1a')
plt.plot(weights_net_rs2, 'm^', label = 'rs2')
plt.plot(weights_net_rs1b, 'g+', label = 'rs1b')
plt.legend()
plt.xlabel('Vectorized weight index')
plt.ylabel('Weight value')

plt.show()
```
